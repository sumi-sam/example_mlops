# **importing and loading**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error,r2_score
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
import xgboost as xgb
```

```python
# Carregando dataset

data = pd.read_csv('/kaggle/input/medical-insurance-payout/expenses.csv')
data.head()
```

```python
# ALTERE OS VALORES PARA PREVISÃO NO FINAL

#Alterar com numero inteiro
age = 25

#Alterar com 'male' ou 'female'
sex = 'male'

#Alterar com imc
bmi = 22.500

#Alterar com numero inteiro
children = 0

#Alterar com 'yes' ou 'no'
smoker = 'no'

#Alterar com 'southeast','southwest','northwest' ou 'northeast'
region = 'northeast'
   
```

```python
if smoker == 'no':
    smoker = 0
if smoker == 'yes':
    smoker = 1
    
if region == 'northeast':
    region_northwest = 0
    region_southeast = 0
    region_southwest = 0
    
if region == 'southeast':
    region_northwest = 0
    region_southeast = 1
    region_southwest = 0

if region == 'southwest':
    region_northwest = 0
    region_southeast = 0
    region_southwest = 1
    
if region == 'northwest':
    region_northwest = 1
    region_southeast = 0
    region_southwest = 0
    
    
if sex == 'male':
    sex = 1
if sex == 'female':
    sex = 0
```

```python
dados = {'age': age, 'bmi': bmi, 'children': children, 'sex_male': sex, 'smoker_yes': smoker,
 'region_northwest': region_northwest, 'region_southeast': region_southeast, 'region_southwest': region_southwest}
pessoa = pd.DataFrame(dados , index= [0])

pessoa
```

# **Transformação dos dados**

```python
#Transformando coluna de objetos em numerico, separando a coluna em 2 e usando valores 1 e 0 para facilitar no modelo

data = pd.get_dummies(data,drop_first=True)
data.head()
```

```python
#verificando nome das colunas

data.columns
```

```python
#Colocando a coluna 'charges' na frente para facilitar quando for separa Variáveis dependentes e independentes

data = data[['charges','age', 'bmi', 'children', 'sex_male', 'smoker_yes',
       'region_northwest', 'region_southeast', 'region_southwest']]
data.columns
```

```python
# Separando variável dependente e independentes

X = data.iloc[:,1:] #variáveis independentes
y = data.iloc[:,0:1] #variável dependente
```

```python
X.head()
```

```python
y.head()
```

```python
# Separando dados de teste e treino

X_train , X_test , y_train , y_test = train_test_split(X,y, test_size=0.2, random_state = 0)
```

# **Construção e avaliação do modelo**

### Regreção linear

```python
# LR = Linear Regression

LR = LinearRegression()
LR = LR.fit(X_train,y_train)

predict = LR.predict(X_test)

print ('Coeficiente de determinação: ',r2_score(y_test,predict))
print ('Erro Quadrático Médio: ',mean_squared_error(y_test,predict))
```

### Randon forest

```python
# RF = Random Forest

RF = RandomForestRegressor()
RF = RF.fit(X_train,y_train)

predict = RF.predict(X_test)

print ('Coeficiente de determinação: ',r2_score(y_test,predict))
print ('Erro Quadrático Médio: ',mean_squared_error(y_test,predict))
```

### XGboost

```python
# xgbR = XGBoost Regressor

xgbR = xgb.XGBRegressor()
xgbR = xgbR.fit(X_train,y_train)

predict = xgbR.predict(X_test)

print ('Coeficiente de determinação: ',r2_score(y_test,predict))
print ('Erro Quadrático Médio: ',mean_squared_error(y_test,predict))
```
